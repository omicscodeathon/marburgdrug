# DATA PROCESSING

```python
import pandas as pd
import numpy as np
import os

raw_file_path = os.path.join('..', 'data', 'raw', 'MARBURG_ASSAY_DATA_RAW.csv')

df = pd.read_csv(raw_file_path)
df
```

```python
import pandas as pd

# Select only the two necessary columns
filtered_df = df[['PUBCHEM_EXT_DATASOURCE_SMILES','Potency', 'PUBCHEM_ACTIVITY_OUTCOME']]
filtered_df = filtered_df[5:]
filtered_df.reset_index(inplace=True,drop=True)


# # Save the new dataset
# filtered_df.to_csv('filtered_data.csv', index=False)
# df = pd.read_csv('filtered_data.csv')
# df
filtered_df.head()
```

```python
# Counting the occurrences of each unique value in the 'PUBCHEM_ACTIVITY_OUTCOME' column
outcome_counts = filtered_df['PUBCHEM_ACTIVITY_OUTCOME'].value_counts()

# Print the result
print(outcome_counts)
```

# DATA REMOVAL

```python
# Remove rows where the outcome is 'inconclusive'
df_cleaned = filtered_df[filtered_df['PUBCHEM_ACTIVITY_OUTCOME'] != 'Inconclusive']
df_cleaned
```

```python
df_cleaned['PUBCHEM_ACTIVITY_OUTCOME'].value_counts()
```

```python
# Randomly sample 7000 inactives
inactives = df_cleaned[df_cleaned['PUBCHEM_ACTIVITY_OUTCOME'] == 'Inactive']
inactives_sampled = inactives.sample(n=7000, random_state=42)  # Set random_state for reproducibility

# Combine the sampled inactives with the actives
actives = df_cleaned[df_cleaned['PUBCHEM_ACTIVITY_OUTCOME'] == 'Active']
df_balanced = pd.concat([actives, inactives_sampled])

# Shuffle the rows to mix actives and inactives
df_new = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
df_new
```

```python
print(df_new['PUBCHEM_ACTIVITY_OUTCOME'].value_counts())
```

# BINARIZTION

```python
# Convert 'Active' to 1 and 'Inactive' to 0
df_new['PUBCHEM_ACTIVITY_OUTCOME'] = df_new['PUBCHEM_ACTIVITY_OUTCOME'].map({'Active': 1, 'Inactive': 0})
df_new = df_new[['PUBCHEM_EXT_DATASOURCE_SMILES', 'PUBCHEM_ACTIVITY_OUTCOME']]
df_new.head()
```

# DATA FILTERING

```python
# Remove rows with NaN values in any column
df_filtered = df_new.dropna()
```

```python
# Remove duplicates
df_final = df_filtered.drop_duplicates()
df_final
```

```python
print(df_final['PUBCHEM_ACTIVITY_OUTCOME'].value_counts())
```

```python
df_final.columns = ['CANONICAL_SMILES', 'ACTIVITY']
df_final.head()
```

```python
df_final.to_csv(os.path.join('..', 'data', 'preprocessed', '001_preprocessed_data.csv'), index=False)
```

# Splitting Dataset

```python
preprocessed_file_path = os.path.join('..', 'data', 'preprocessed', '001_preprocessed_data.csv')
data = pd.read_csv(preprocessed_file_path)
X = data['CANONICAL_SMILES'].copy()
y = data['ACTIVITY'].copy()
data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 17)
```

```python
train_dataset = pd.concat([X_train, y_train], axis=1)
train_dataset.head()
```

```python
test_dataset = pd.concat([X_test, y_test], axis = 1)
test_dataset.head()
```

```python
train_dataset.to_csv(os.path.join('..', 'data', 'preprocessed', '001_preprocessed_train_data.csv'), index=False)
test_dataset.to_csv(os.path.join('..', 'data', 'preprocessed', '001_preprocessed_test_data.csv'), index=False)
```
